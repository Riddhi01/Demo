```python
import pandas as pd
# Sample DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Age': [25, 30, 35]})
# Merge DataFrames on 'ID'
merged_df = pd.merge(df1, df2, on='ID')
print(merged_df)
```

```python
# Sample DataFrames
df1 = pd.DataFrame({'Employee': ['John', 'Emma'], 'Department': ['HR', 'Finance']})
df2 = pd.DataFrame({'Employee': ['John', 'Emma'], 'Score': [85, 90]})

# Concatenate vertically
df_vertical = pd.concat([df1, df2], axis=0, ignore_index=True)
print(df_vertical)
```

```python
# Concatenate horizontally
df_horizontal = pd.concat([df1, df2], axis=1)
print(df_horizontal)
```

```python
# Create sample DataFrames
customers = pd.DataFrame({'CustomerID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
transactions = pd.DataFrame({'CustomerID': [1, 2, 2, 3], 'Amount': [100, 200, 150, 300]})

# Merge DataFrames
merged_df = pd.merge(customers, transactions, on='CustomerID')
print(merged_df)
```

```python
# Group by CustomerID and calculate total amount
total_amount = merged_df.groupby('Name')['Amount'].sum()
print(total_amount)
```

```python
# Sample DataFrame
sales_data = pd.DataFrame({
    'Product': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'SalesAmount': [100, 200, 150, 250, 120, 300]
})
print(sales_data)
```

```python
# Group by Product and Region and calculate sum and mean
grouped_sales = sales_data.groupby(['Product', 'Region'])['SalesAmount'].agg(['sum', 'mean'])
print(grouped_sales)
```
